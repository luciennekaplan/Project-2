```python
import pandas as pd
import re
```

```python
df = pd.read_csv("data/BusinessAnalyst.csv")
df
```

```python
df = df.drop(columns=["index", "Unnamed: 0"])
df
```

```python
numbers = []
for word in df["Salary Estimate"]:
    numbers.append(re.findall(r'\d+', word))


print(numbers)
```

```python
high = []
low = []
index = 0
for listy in numbers:
    low.append(int(listy[0])* 1000)
    try:
        high.append(int(listy[1])*1000)
    except IndexError:
        high.append("null")
    index += 1
print(high)
print(low)
```

```python
df["Salary Low"] = low
df["Salary High"] = high
df.head()
```

```python
df.loc[df["Salary High"] == "null", :]
```

```python
df = df.loc[df["Salary High"] != "null", :]
df
```

```python
df["Size"].value_counts()
```

```python
df['Size'] = df['Size'].str.replace('-1', 'Unknown')
```

```python
df['Industry'] = df['Size'].str.replace('-1', 'Other')
```

```python
df['Company Name'] = df['Company Name'].str.replace('\\n', '')
```

```python
df['Company Name'] = df['Company Name'].str.replace('.', '')
```

```python
df['Company Name'] = df['Company Name'].str.replace('\d+', '')
df
```

```python
df.to_csv("data/CleanBusinessAnalyst.csv", index = False)
```
